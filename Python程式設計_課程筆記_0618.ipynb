````python
# 1700000000
# Tue Nov 14 2023 22:13:20 GMT+0000

input_second = 1700000000

# ÊØèÂñÆ‰ΩçÁöÑÁßíÊï∏ÊèõÁÆó
second_per_hr = 60 * 60
second_per_day = 60 * 60 * 24

# Ë®àÁÆóÊúâÂπæÂ§©
day = input_second // second_per_day # Â§©Êï∏ == Á∏ΩÁßíÊï∏Ë∑üÊØèÂ§©ÊúâÂπæÁßíÔºåÂèñÂïÜÔºàÂèØ‰ª•Êâ£ÊéâÂπæÂ§©Â∞±Á≠âÊñºÊúâÂπæÂ§©Ôºâ
remain_second = input_second % second_per_day # Ââ©È§òÁöÑÁßíÊï∏ == Á∏ΩÁßíÊï∏ÔºåÂèñÈ§òÊï∏ÔºàÈô§Ââ©‰∏ã‰æÜÁöÑÊï∏Â≠óÔºåÂ∞±ÊòØÂâ©È§òÁßíÊï∏ÂõâÔºâ

# Ë®àÁÆóÂπæÂ∞èÊôÇ
hour = remain_second // second_per_hr # Â∞èÊôÇ == Â∞ç(Èô§ÂÆåÂ§©Êï∏ÁöÑ)Ââ©È§òÁßíÊï∏ÂèñÂïÜ
remain_second = remain_second % second_per_hr

# Ë®àÁÆóÂπæÂàÜÈêò
min = remain_second // 60 # ÂàÜÈêò == Â∞ç(Èô§ÂÆåÂ∞èÊôÇÁöÑ)Ââ©È§òÁßíÊï∏ÂèñÂïÜ

# Ââ©‰∏ãÁöÑÂ∞±ÊòØÁßí
sec = remain_second % 60 # ÁßíÊï∏ == Â∞ç(Èô§ÂÆåÂ∞èÊôÇÁöÑ)Ââ©È§òÁßíÊï∏ÂèñÈ§òÊï∏

# ÂàÜÈñãÂç∞
print("Â§©Ôºö", day)
print("Â∞èÊôÇÔºö", hour)
print("ÂàÜÈêòÔºö", min)
print("ÁßíÔºö", sec)

# Âç∞Âá∫‰∏ÄÊï¥Ë°å
print(f"{input_second} Áßí = {day} Â§© {hour} Â∞èÊôÇ {min} ÂàÜÈêò {sec} Áßí")
````
